##### 4.4.8 MUTIVARIATE ANALYSIS (How do lunch type and test preparation course completion affect students' performance?)
Interaction Between Lunch Type and Test Prep:

```python
# Set Seaborn theme and color palette
sns.set_theme(style="whitegrid", palette="Set2")

plt.figure(figsize=(14, 8))

# Subplot 1: Math Score
plt.subplot(2, 2, 1)
sns.barplot(x='lunch', y='math_score', hue='test_preparation_course', data=df)
plt.title("Math Score by Lunch Type and Test Preparation")
plt.xlabel("Lunch Type")
plt.ylabel("Math Score")

# Subplot 2: Reading Score
plt.subplot(2, 2, 2)
sns.barplot(x='lunch', y='reading_score', hue='test_preparation_course', data=df)
plt.title("Reading Score by Lunch Type and Test Preparation")
plt.xlabel("Lunch Type")
plt.ylabel("Reading Score")

# Subplot 3: Writing Score
plt.subplot(2, 2, 3)
sns.barplot(x='lunch', y='writing_score', hue='test_preparation_course', data=df)
plt.title("Writing Score by Lunch Type and Test Preparation")
plt.xlabel("Lunch Type")
plt.ylabel("Writing Score")

# Optional: remove unused 4th subplot or use it
plt.subplot(2, 2, 4)
plt.axis('off')  # You can remove or use this space

plt.tight_layout()
plt.show()
```

##### 4.4.9 MUTIVARIATE ANALYSIS (How is parental level of education related to students' test preparation and socioeconomic status?)
- Are students more likely to complete a test preparation course based on their parents' level of education?
    - This helps identify if parental education influences a student's likelihood of engaging in academic preparation.

- How does parental education level relate to the type of lunch students receive (standard vs free/reduced)?
    - This explores the link between parental education and the student’s socioeconomic background (as inferred by lunch type).

```python

# Set style and color palette
sns.set_theme(style="whitegrid", palette="Set2")

# Create the figure and axes
f, ax = plt.subplots(1, 2, figsize=(20, 8))

# Plot 1: Test preparation course vs Parental education
sns.countplot(
    x='parental_level_of_education',
    hue='test_preparation_course',
    data=df,
    palette='Set2',
    saturation=0.95,
    ax=ax[0]
)
ax[0].set_title('Test Preparation by Parental Education', fontsize=20, color='black')
ax[0].set_xlabel('Parental Level of Education', fontsize=14)
ax[0].set_ylabel('Number of Students', fontsize=14)
ax[0].tick_params(axis='x', rotation=25)
ax[0].legend(title='Test Preparation', fontsize=12, title_fontsize=13)

for container in ax[0].containers:
    ax[0].bar_label(container, fontsize=12, color='black')

# Plot 2: Lunch type vs Parental education
sns.countplot(
    x='parental_level_of_education',
    hue='lunch',
    data=df,
    palette='Set3',
    saturation=0.95,
    ax=ax[1]
)
ax[1].set_title('Lunch Type by Parental Education', fontsize=20, color='black')
ax[1].set_xlabel('Parental Level of Education', fontsize=14)
ax[1].set_ylabel('Number of Students', fontsize=14)
ax[1].tick_params(axis='x', rotation=25)
ax[1].legend(title='Lunch Type', fontsize=12, title_fontsize=13)

for container in ax[1].containers:
    ax[1].bar_label(container, fontsize=12, color='black')

# Improve layout
plt.tight_layout()
plt.show()
```

Insights:
- Higher parental education correlates with more test preparation:
    - Students whose parents have bachelor’s or master’s degrees are more likely to complete the test preparation course.
    - This suggests that more educated parents may prioritize or have better access to academic support programs for their children.

- Lower parental education corresponds with less preparation:
    - Students with parents who have only high school or some high school education show lower participation in test prep.
    - This might reflect lack of resources, awareness, or encouragement at home.

- Parental education strongly correlates with lunch type:
    - Students whose parents have lower levels of education are more likely to receive free/reduced lunch, indicating a lower socioeconomic status.
    - Conversely, those whose parents have higher education tend to have standard lunch, implying a more stable economic background.

- Socioeconomic disparities are apparent:
    - The contrast in lunch types by parental education level reveals a clear socioeconomic divide, which may also influence academic outcomes.

Overall Implications:
- Parental education level appears to influence both academic preparation and socioeconomic status.
- Students from more advantaged backgrounds (educated parents and standard lunch) may have better academic support systems.
- These factors can contribute to educational inequality and may explain differences in student performance across groups.
